```python
import pandas as pd
df=pd.read_csv('customer_shopping_behavior.csv')  #imiporting csv file into pandas dataframe
```

```python
 #checking missing values per column.
#here reveiw rating columns has null value ,  decide whether to drop or impute missing data.
#Nulls are ignored in aggregations in pythong,
#if you use a column with nulls without handling them, you’ll often get NaN results or lose rows in filtering.
df.isna().sum()   
```

```python
# filling null(na value with median, which recommeded, because mean(average) will create outliers
#df.dropna() → remove rows/columns with nulls , our analysis may go wrong, 
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x:x.fillna(x.median())) 
#above line fills missing review ratings with the median rating of their category,
#ensuring more meaningful imputation than just using a global median.
#Median is robust against outliers, making it a good choice for imputation
```

```python
df.isnull().sum()    #checkig for na column filled with median value or not
```

```python
# if you observer columns names are with caps and small letters and spaces bewteen words, its very difficult to handle,
#lets make it as all small letter and one word
df.columns=df.columns.str.lower()   
df.columns=df.columns.str.replace(' ','_')  # repalceing space with _
 # purchase amount column format doese not look good, and diffuclt to handle, so renaming column name
df=df.rename(columns={'purchase_amount_(usd)':'purchase_amount'}) 
```

```python
#lets check columns name renamed and got as expected or not
df.columns
```

```python
#creating new column  as age_group based on age
#lets label age group as below by creating list 
labels=['Young Adult','Adult','Middle Aged','Senior']
#categorize continuous variables (like age) into quantile-based groups.
#createing an age group column by splitting ages into 4 equally sized bins and labeling them with your custom names from label variable.
df['age_group']=pd.qcut(df['age'],q=4,labels = labels) 
```

```python
#lets check  new colum with 10 sample data
df[['age','age_group']].head(10)
```

```python
#lets observer frequency_of_purchases columns which has categorical text data, which we can cant campare with each other columns
#lets see those data,by printing unique data
print(df['frequency_of_purchases'].unique())
```

```python
#lets create new column purchase_frequecy_days with numeric column
#to create new column we need numberic values, so lets create list variable as frequency_mapping , this we can map later
frequency_mapping={
    'Fortnightly':14,
    'Weekly':7,
    'Annually':365,
    'Quarterly':90,
    'Bi-Weekly':14,
    'Monthly':30,
    'Every 3 Months':90
}
#create new columns with nubmers by mappig above list
df['purchase_frequecy_days']=df['frequency_of_purchases'].map(frequency_mapping) 
```

```python
# lets check mapped numbers with text for those coumns with 10 rows
df[['frequency_of_purchases','purchase_frequecy_days']].head(10)
```
